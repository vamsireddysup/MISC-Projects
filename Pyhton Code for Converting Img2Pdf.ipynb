```python
import os
from PIL import Image

def convert_subfolders_to_pdfs(root_folder, output_folder):
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for dirpath, dirnames, filenames in os.walk(root_folder):
        # Skip the root folder itself, only process subfolders
        if dirpath == root_folder:
            continue

        images = []
        for filename in sorted(filenames):
            if filename.lower().endswith(image_extensions):
                images.append(os.path.join(dirpath, filename))

        if images:
            # Prepare images
            image_list = []
            for file in images:
                img = Image.open(file).convert('RGB')
                image_list.append(img)

            # Prepare output PDF name based on subfolder name
            subfolder_name = os.path.basename(dirpath)
            output_pdf_path = os.path.join(output_folder, f"{subfolder_name}.pdf")

            # Save images into a PDF
            first_image = image_list[0]
            remaining_images = image_list[1:]

            first_image.save(output_pdf_path, save_all=True, append_images=remaining_images)
            print(f"Generated PDF for {subfolder_name} at {output_pdf_path}")

# Example usage
root_images_folder = "/home/vamsidhar-reddy-eraganeni/Desktop/Waste/UVM_Lucid/UVM/UVM slides"
output_pdfs_folder = "/home/vamsidhar-reddy-eraganeni/Desktop/Waste/UVM_Lucid/UVM/UVM slides/UVM_subfolder_pdfs"

convert_subfolders_to_pdfs(root_images_folder, output_pdfs_folder)
```
